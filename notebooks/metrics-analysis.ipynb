```python
# get latest champ
playoff_brackets = pd.read_sql_table('playoff_brackets', engine)
playoff_brackets.sort_values('id', ascending=False, inplace=True)
simulation_id = playoff_brackets.iloc[0]['simulation_id']
simulation_year = playoff_brackets.iloc[0]['year']
nodes_sep_comma = playoff_brackets.iloc[0]['nodes_sep_comma']
```

```python
champion_team_id = nodes_sep_comma.split(',')[0]
champion_team_id
```

```python
# finals
pd.read_sql_query('''select g.*
                            ,gs.*
                            , p.* 
                            from games g
                            join game_stats gs on gs.game_id = g.id
                            join players p on gs.player_id = p.id
                          where 1=1
                          and g.game_type = 'FINALS'
                          and g.simulation_id = '{simulation_id}'
                          and g.year = {year}
                          and (home_team_id = {team_id} or away_team_id = {team_id})
                          and p.team_id = {team_id}
                          '''.format(
    team_id=champion_team_id, year=simulation_year, simulation_id=simulation_id),
                          engine)
```

```python
games = pd.read_sql_query('''select g.*
                            ,gs.*
                            , p.* 
                            from games g
                            join game_stats gs on gs.game_id = g.id
                            join players p on gs.player_id = p.id
                          where 1=1
                          --and g.game_type = 'FINALS'
                          and g.simulation_id = '{simulation_id}'
                          and g.year = {year}
                          and (home_team_id = {team_id} or away_team_id = {team_id})
                          and p.team_id = {team_id}
                          '''.format(
    team_id=champion_team_id, year=simulation_year, simulation_id=simulation_id),
                          engine)
```

```python
games.groupby(['player_id','team_id'], as_index=False).agg(
    {'points_scored': 'mean', 'name': 'first', 'year_drafted':'first', 'points_per_game': 'first'
                                               }).sort_values('points_per_game', ascending=False).head(30)
```

## Metrics

```python
# number of players scoring 20ppg or more
```

```python
grouped_players = games.groupby(['player_id','team_id', 'year'], as_index=False).agg(
    {'points_scored': 'mean', 'name': 'first', 'year_drafted':'first', 'points_per_game': 'first'
                                               }).sort_values('points_per_game', ascending=False)
grouped_players['above_20ppg'] = grouped_players['points_scored'].apply(lambda x: 1 if x > 20. else 0)
grouped_players.groupby(['team_id'], as_index=False).agg({'above_20ppg': 'sum'}).head()
#rouped_players.head()
```
